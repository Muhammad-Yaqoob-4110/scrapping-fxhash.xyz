```python
libraries_list = df['Generative Library'].tolist()
```

```python
from openai import OpenAI
```

```python
from google.colab import userdata
NVIDIA_API_KEY = userdata.get('NVIDIA_API_KEY')
```

```python
client = OpenAI(
  base_url = "https://integrate.api.nvidia.com/v1",
  api_key = NVIDIA_API_KEY
)
```

```python
def make_prompt(libraries_list):
    formatted_list = [f"<div> {text} </div>" for text in libraries_list]
    prompt = f"""
          I have a list of texts of length 10. Each text starts with `<div>` and ends with `</div>`.
          Your task is to identify if any JavaScript creative coding library (e.g., p5.js, Three.js) is mentioned in each text.
          If a library is mentioned, return just the name of the library. If no library is mentioned, return 'Unknown'.

          Please return exactly one result for each text, corresponding to each item in the list.
          There should be a one-to-one mapping: one result (library name or 'Unknown') for each text, in the same order as provided.

          The texts are:
          {', '.join(formatted_list)}

          Return the results as a comma-separated, where each item corresponds to a single text.
          Ensure that the length of the returned list matches the number of texts provided means must be 10.
          """
    return prompt
```

```python
lib_list = []
```

```python
for lib in libraries_list:
  prompt = f"""
          Your task is to identify if any JavaScript creative coding library (e.g., p5.js, Three.js) is mentioned in each text.
          If a library is mentioned, return just the name of the library. If no library is mentioned, return 'Unknown'.

          Here is the {lib}
          """
  completion = client.chat.completions.create(
    model="meta/llama-3.1-8b-instruct",
    messages=[{"role":"user","content": prompt}],
    temperature=0.2,
    top_p=0.7,
    max_tokens=1024,
    stream=True
    )

  full_response = ""
  for chunk in completion:
    if chunk.choices[0].delta.content is not None:
      full_response += chunk.choices[0].delta.content
  lib_list.append(full_response)
  print(full_response)
```

```python
if len(lib_list) < len(df):
    # Extend lib_list with empty strings
    lib_list.extend([''] * (len(df) - len(lib_list)))
elif len(lib_list) > len(df):
    # Trim lib_list to match the length of the DataFrame if it's too long
    lib_list = lib_list[:len(df)]

# Add lib_list as a new column
df['Library Mentioned'] = lib_list
```

```python
df.to_csv("/content/fxhash_data_with_library.csv", index=False)

print("Updated DataFrame saved to fxhash_data_with_library.csv")
```
